Now,we read the train and test CSV files, combine the `title` and `text` into a single `text` column, removes empty entries, and adjust label values if they exceed 13.


```python
import pandas as pd
train_df = pd.read_csv(TRAIN_PATH, header=None, names=["label","title","text"])
test_df  = pd.read_csv(TEST_PATH,  header=None, names=["label","title","text"])

train_df["text"] = train_df["title"].fillna("").astype(str) + ". " + train_df["text"].fillna("").astype(str)
test_df["text"]  = test_df["title"].fillna("").astype(str) + ". " + test_df["text"].fillna("").astype(str)

train_df = train_df[train_df["text"].str.strip().astype(bool)]
test_df  = test_df[test_df["text"].str.strip().astype(bool)]

if train_df["label"].max() > 13:
    train_df["label"] -= 1
    test_df["label"]  -= 1

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
```

This code splits the training data into training and validation sets (90% train, 10% validation) while keeping the class distribution balanced using stratif


```python
from sklearn.model_selection import train_test_split

train_texts, val_texts, train_labels, val_labels = train_test_split(
    train_df["text"], train_df["label"], test_size=0.1, stratify=train_df["label"], random_state=42
)

print("Train:", len(train_texts), "Val:", len(val_texts))
```

The below code ensures stratified sampling preserved class balance, preventing unintended bias.

```python
# Verify class balance is preserved after stratified split
original_dist = train_df["label"].value_counts(normalize=True).sort_index()
split_dist = train_labels.value_counts(normalize=True).sort_index()

diff = abs(original_dist - split_dist)

print("Maximum distribution difference:", diff.max())

assert all(diff < 0.01), "Class distribution shifted after split."

print("✔ Stratified split preserved class balance within 1% tolerance.")
```

This code visualizes the number of samples per class in the training dataset using a Seaborn count plot, labeling each bar with its corresponding category name to confirm the dataset is balanced across


```python
import seaborn as sns
import matplotlib.pyplot as plt

class_names = []
with open(os.path.join(DRIVE_ROOT, "classes.txt"), "r") as f:
    class_names = [line.strip() for line in f if line.strip()]

plt.figure(figsize=(12,6))
ax = sns.countplot(x=train_df["label"], palette="Blues")
plt.title("DBpedia Dataset: Balanced Class Distribution Across 14 Categories", fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Class Label", fontsize=12, fontweight='bold')
plt.ylabel("Number of Samples", fontsize=12, fontweight='bold')
plt.xticks(range(14), [f"{class_names[i]}" if i < len(class_names) else str(i) for i in range(14)], rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3, linestyle='--')

plt.show()
```

This code calculates the number of words in each text and displays summary statistics. It also plots a histogram of text lengths with a KDE curve, showing the distribution and key statistics (mean, median, min,


```python
train_df["len_words"] = train_df["text"].str.split().apply(len)
print(train_df["len_words"].describe())

plt.figure(figsize=(12,6))
ax = sns.histplot(train_df["len_words"], bins=60, kde=True, color='steelblue', edgecolor='black', linewidth=0.5)
plt.xlim(0, 250)
plt.title("DBpedia Dataset: Text Length Distribution (Words per Sample)",
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Text Length (Number of Words)", fontsize=12, fontweight='bold')
plt.ylabel("Frequency (Number of Samples)", fontsize=12, fontweight='bold')

# Custom Legend
from matplotlib.patches import Patch
legend_elements = [
    f'Mean: {train_df["len_words"].mean():.1f} words',
    f'Median: {train_df["len_words"].median():.0f} words',
    f'Min: {train_df["len_words"].min():.0f} words',
    f'Max: {train_df["len_words"].max():.0f} words'
]
plt.legend(handles=[Patch(facecolor='none', edgecolor='none', label=label) for label in legend_elements],
           loc='upper right', fontsize=10, framealpha=0.9, handlelength=0, handletextpad=0)

plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.show()
```

Although the dataset appears balanced numerically, average text length varies across classes. This introduces structural bias: models may learn length-based heuristics instead of true semantic understanding.

```python
# Average text length per class
length_by_class = train_df.groupby("label")["len_words"].mean().sort_values()

print("Average Text Length Per Class:\n")
print(length_by_class)

plt.figure(figsize=(12,6))
length_by_class.plot(kind='bar')
plt.title("Average Text Length per Class (Bias Analysis)",
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Class Label", fontsize=12, fontweight='bold')
plt.ylabel("Average Words per Sample", fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.show()

print(f"\nLength disparity ratio (max/min): {length_by_class.max() / length_by_class.min():.2f}")
```

Observed differences in vocabulary richness across classes suggest uneven lexical diversity. Transformer-based models may disproportionately favor linguistically richer categories.

```python
from sklearn.feature_extraction.text import CountVectorizer

vocab_sizes = {}

for label in train_df["label"].unique():
    subset = train_df[train_df["label"] == label]["text"]
    vectorizer = CountVectorizer(stop_words="english", max_features=5000)
    X = vectorizer.fit_transform(subset)
    vocab_sizes[label] = len(vectorizer.vocabulary_)

vocab_series = pd.Series(vocab_sizes).sort_values()

print("\nVocabulary Size per Class:\n")
print(vocab_series)

plt.figure(figsize=(12,6))
vocab_series.plot(kind='bar')
plt.title("Vocabulary Richness per Class",
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Class Label", fontsize=12, fontweight='bold')
plt.ylabel("Unique Vocabulary Size", fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.show()

print(f"\nVocabulary disparity ratio (max/min): {vocab_series.max() / vocab_series.min():.2f}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Limit vocabulary size to keep computation manageable for EDA
vectorizer = TfidfVectorizer(max_features=5000)

X = vectorizer.fit_transform(train_df["text"])

# Compute sparsity ratio
sparsity = 1 - (X.nnz / (X.shape[0] * X.shape[1]))

print("TF-IDF Matrix Shape:", X.shape)
print("Number of Non-Zero Entries:", X.nnz)
print("Sparsity:", round(sparsity, 4))
```

**Interpretation**:

The matrix has shape (560,000 × 5,000).

Despite the high dimensionality, the sparsity is extremely high (close to 1).

This confirms that TF-IDF produces a sparse representation.

Sparse linear models (e.g., Logistic Regression, Linear SVM) remain computationally feasible on a single machine.

However, if the vocabulary were expanded to 100K+ features, memory usage and training time would increase substantially.

**Implication:**

Classical bag-of-words approaches are feasible and scalable.

Transformer-based methods, while semantically richer, require GPU acceleration due to dense representations and quadratic attention complexity.

This directly informs the comparison between course techniques (TF-IDF) and beyond-course techniques (transformers).

```python
# ==============================
# EDA Validation & Sanity Checks
# ==============================

print("Running EDA validation checks...\n")

# 1. Missing values check
assert train_df.isnull().sum().sum() == 0, "Missing values detected."
print("✔ No missing values.")

# 2. Duplicate rows check
assert train_df.duplicated().sum() == 0, "Duplicate rows detected."
print("✔ No duplicate rows.")

# 3. Label integrity check (DBpedia: 14 classes labeled 0–13)
expected_labels = set(range(14))
actual_labels = set(train_df["label"].unique())
assert actual_labels == expected_labels, "Unexpected label values detected."
print("✔ Label set verified (14 expected classes present).")

# 4. Empty document check
empty_docs = (train_df["text"].str.strip() == "").sum()
assert empty_docs == 0, "Empty documents detected."
print("✔ No empty documents after preprocessing.")

# 5. TF-IDF matrix consistency check
assert X.shape[0] == train_df.shape[0], "Mismatch between TF-IDF rows and dataset size."
assert X.shape[1] <= 5000, "TF-IDF exceeded specified max_features."
print("✔ TF-IDF feature matrix dimensions verified.")

# 6. Sparsity sanity check
sparsity = 1 - (X.nnz / (X.shape[0] * X.shape[1]))
assert sparsity > 0.90, "Matrix is unexpectedly dense."
print(f"✔ Sparse representation confirmed (sparsity = {round(sparsity, 4)}).")

print("\nAll EDA validation checks passed successfully.")
```
